```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/used_cars_dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
print(df['kmDriven'].iloc[0])
df['kmDriven'] = df['kmDriven'].astype(str)
df['kmDriven'] = df['kmDriven'].str.replace(' km', '')
df['kmDriven'] = df['kmDriven'].str.replace(',', '')
df['kmDriven'] = pd.to_numeric(df['kmDriven'], errors='coerce')
print(df['AskPrice'].iloc[0])

df['AskPrice'] = df['AskPrice'].astype(str)
df['AskPrice'] = df['AskPrice'].str.replace('₹', '')
df['AskPrice'] = df['AskPrice'].str.replace(',', '')
df['AskPrice'] = df['AskPrice'].str.strip() 
df['AskPrice'] = pd.to_numeric(df['AskPrice'], errors='coerce')
df[['kmDriven', 'AskPrice']].head()
```

```python
df['AdditionInfo'].head(10)
```

```python
known_brands = df['Brand'].dropna().unique()
print(known_brands)
```

```python
def find_brand(text):
    for brand in known_brands:
        if brand.lower() in str(text).lower():
            return brand
    return np.nan

missing_brand_mask = df['Brand'].isnull()
df.loc[missing_brand_mask, 'Brand'] = df.loc[missing_brand_mask, 'AdditionInfo'].apply(find_brand)

df['Brand'].isnull().sum()
```

```python
known_models = df['model'].dropna().unique()
print(known_models[:30])
```

```python
def find_model(text):
    for model in known_models:
        if model.lower() in str(text).lower():
            return model
    return np.nan

missing_model_mask = df['model'].isnull()
df.loc[missing_model_mask, 'model'] = df.loc[missing_model_mask, 'AdditionInfo'].apply(find_model)
df['model'].isnull().sum() 
```

```python
df.isnull().sum()
```

```python
df.loc[df['Year'].isnull() & df['Age'].notnull(), 'Year'] = 2024 - df['Age']
df.loc[df['Age'].isnull() & df['Year'].notnull(), 'Age'] = 2024 - df['Year']
df[['Year', 'Age']].isnull().sum()
```

```python
df['Year'] = df['Year'].fillna(df['Year'].median())
df['Age'] = df['Age'].fillna(df['Age'].median())
df[['Year', 'Age']].isnull().sum()
```

```python
df['kmDriven'] = df['kmDriven'].fillna(df['kmDriven'].median())
df['kmDriven'].isnull().sum()
```

```python
df['Transmission'] = df['Transmission'].fillna(df['Transmission'].mode()[0])
df['Owner'] = df['Owner'].fillna(df['Owner'].mode()[0])
df['FuelType'] = df['FuelType'].fillna(df['FuelType'].mode()[0])
df['PostedDate'] = df['PostedDate'].fillna(df['PostedDate'].mode()[0])

df[['Transmission', 'Owner', 'FuelType', 'PostedDate']].isnull().sum()
```

```python
df = df.dropna(subset=['Brand', 'model'])
df.isnull().sum()
```
